```python
mean = 5
std_dev = 1.5
```

```python
z1= (3.17 - mean) / std_dev
p1 = stats.norm.cdf(z1)
p1 = round(p1,3)
print(p1)
```

```python
x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 1000)
y = stats.norm.pdf(x,mean,std_dev)
plt.figure(figsize=(9,5))
plt.plot(x,y, color='brown')
x_fill = np.linspace(mean - 3*std_dev, 3.17, 1000)
y_fill = stats.norm.pdf(x_fill, mean, std_dev)
plt.fill_between(x_fill, y_fill, alpha = 0.5, label='Breaking strength < 3.17Kg')
plt.title('proportion of gunny bags Breaking strength less than 3.17',pad=20)
plt.xlabel('Breaking strength (kg/sq cm)',labelpad = 15)
plt.ylabel('Density', labelpad = 15)
plt.legend(loc='upper right');
```
